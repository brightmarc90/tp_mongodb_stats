### Calculez l'écart type des montants des transactions pour comprendre la variabilité des ventes.

```python
import statistics as stats
moyenne = stats.mean(df["total_amount"])
print(f"Moyenne: {moyenne}")
ecartType = stats.pstdev(df["total_amount"])
print(f"Ecart type: {ecartType}")
```

les ventes sont dispersées autour de la moyenne donc un ecart type élevé.
les données montrent une grande variabilité.

### Calculez la médiane des montants des transactions pour chaque magasin.

```python
medianes = pd.pivot_table(df, values="total_amount", index="store_id", aggfunc="median")
medianes
```

### Déterminez les quartiles pour les montants des transactions afin de mieux comprendre la distribution des ventes.

```python
q1 = df["total_amount"].quantile(0.25)
q2 = df["total_amount"].quantile(0.50)
q3 = df["total_amount"].quantile(0.75)
print(f"quantile 1: {q1}")
print(f"quantile 2: {q2}")
print(f"quantile 3: {q3}")
```

L'écart entre Q1 et Q3 est grand. Ceci prouve encore qu'il y a une grande variabilité dans les données de vente.

### Cette question est libre utilisez des graphiques de votre choix pour analyser le dataset.

#### Répartition des transactions de ventes (pie chart)

```python
nbTransByStore = pd.pivot_table(df, index="store_id", aggfunc='size')
proportions = (nbTransByStore / nbTransByStore.sum())*100
plt.figure(figsize=(6, 6))
proportions.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Répartition des transactions par magasin')
plt.show()
```

#### Répartition du niveau des ventes (Chiffre d'affaires) par magasin (bar chart)

```python
totalSalesByStore = pd.pivot_table(df, values="total_amount", index="store_id", aggfunc='sum')
totalSalesByStore.plot(kind="bar", stacked=True)
plt.title("Répartition du chiffre d'affaires par magasin")
plt.ylabel('Montants totaux')
```

#### Visualisation des meilleurs jours par magasin

```python
totalSalesByStore = pd.pivot_table(df, values="total_amount", index=["store_id", "date"], aggfunc='sum')
totalSalesByStore.plot(kind="bar", stacked=True)
plt.title("Meilleurs jours de vente par magasin")
```

#### Evolution du chiffre d'affaires global

```python
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['total_amount'], marker='o', linestyle='-.', color='b')
plt.axhline(moyenne, color='r', linestyle='-', label=f'Moyenne: {moyenne:.2f}')
plt.fill_between(df['date'], moyenne - ecartType, moyenne + ecartType, color='yellow', alpha=0.3, label=f'Ecart type: Â±{ecartType:.2f}')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Montant total')
plt.title('Montant des transactions au fil du temps')
plt.legend()
plt.tight_layout()
plt.show()
```
